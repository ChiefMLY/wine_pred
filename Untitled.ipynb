## Data Preprocessing

```python
#Label Binarization
y = wine_df['quality'].apply(lambda y_val: 1 if y_val >= 7 else 0)
y
```

```python
#Creating the features df
X = wine_df.drop(columns = 'quality', axis = 1)
```

```python
#Splitting the data into training and test data
SEED = 24
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, stratify = y, random_state = SEED)
```

```python
#Validating the split
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

## Model Training

```python
#Initializing a RandomForestClassifier

classifier = RandomForestClassifier()

#Fiitting the model 

classifier.fit(X_train, y_train)
```

## Model Evaluation

```python
#Predicting labels for the test features
y_pred = classifier.predict(X_test)

#Accuracy of the model on the test data
test_acc = accuracy_score(y_pred, y_test)

print(f"The accuracy score of the model on the test data is: {test_acc}")
```

## Building a Predictive System

```python
#Passing in a known row value from our dataframe to the model to label
input_data = (7.4,0.7,0.0,1.9,0.076,11.0,34.0,0.9978,3.51,0.56,9.4)


#Converting the input data into a numpy array
input_array = np.asarray(input_data)

#Reshaping the data because we are only predicting the label for one instance

input_reshaped = input_array.reshape(1,-1)

#Making the prediction 

pred = classifier.predict(input_reshaped)

print(pred)

if (pred[0] == 1):
    print('Good Quality Wine')
else:
    print('Bad Quality Wine')
```
